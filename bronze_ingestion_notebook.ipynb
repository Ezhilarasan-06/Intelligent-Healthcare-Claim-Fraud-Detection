```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%pip install pypdf
```

```python
cost_df = spark.read.csv("/Volumes/catalog1/default/adls/Healthcare project/standard_costs.csv", header=True, inferSchema=True)
```

```python
cost_df.display()
```

```python
policy_path = "dbfs:/Volumes/catalog1/default/adls/Healthcare project/policies/"

policy_binary_df = (
    spark.read.format("binaryFile")
    .load(policy_path)
)

policy_binary_df.display()
```

```python
claim_path = "dbfs:/Volumes/catalog1/default/adls/Healthcare project/claims/"

claims_binary_df = (
    spark.read.format("binaryFile")
    .load(claim_path)
)

claims_binary_df.display()
```

```python
%sql
create database if not exists catalog1.bronze_layer;
```

```python
cost_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("catalog1.bronze_layer.bronze_cost_table")
```

```python
policy_binary_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("catalog1.bronze_layer.bronze_policy_table")
```

```python
claims_binary_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("catalog1.bronze_layer.bronze_claims_table")
```
